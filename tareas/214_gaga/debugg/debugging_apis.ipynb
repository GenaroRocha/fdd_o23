## Pokemon Data

Para  el tipo de pokemon debes obtener:
1. `pokemon_id`
2. `pokemon_name`
3. `type`(s) cada pokemon tiene a 1 o dos tipos (no mas). Tienes que guardar ambos  

Guarda los `moves` del pokemon pues los usaremos mas adelante. En especial el `name` y la `url` de cada movimiento apra que podamos iterar despues

```python
# Elige tu pokemon
pokemon_name = 'lucario'
```

```python
import requests

# Solicitud para obtener datos de un Pokémon
pokemon_name = 'lucario'
pokemon_url = f'https://pokeapi.co/api/v2/pokemon/{pokemon_name}'

pokemon_response = requests.get(pokemon_url)

if pokemon_response.status_code == 200:
    pokemon_data = pokemon_response.json()
    print(f"Nombre del Pokémon: {pokemon_data['name']}")
    print(f"ID del Pokémon: {pokemon_data['id']}")
    print(f"Tipos del Pokémon: {', '.join([type['type']['name'] for type in pokemon_data['types']])}")
    moves = [{'name': move['move']['name'], 'url': move['move']['url']} for move in pokemon_data['moves']]
    print(f"Movimientos del Pokémon: {', '.join([move['name'] for move in moves])}")

    
else:
    print(f"No se pudo obtener la información del Pokémon. Código de respuesta: {pokemon_response.status_code}")
```

## Moves Data

Ahora para cada movimiento `move` del pokemon debes obtener los siguientes datos usando la `url`
1. `name`
2. `accuracy`
3. `type`
4. `power`
5. `pp`

```python

#Ahora para cada movimiento en el diccionario moves, vamos a hacer una solicitud a la API de PokéAPI para obtener la información de cada movimiento:
for move in moves:
    move_response = requests.get(move['url'])
    if move_response.status_code == 200:
        move_data = move_response.json()
        print(f"Nombre del movimiento: {move_data['name']}")
        print(f"Precisión del movimiento: {move_data['accuracy']}")
        print(f"Tipo del movimiento: {move_data['type']['name']}")
        print(f"Poder del movimiento: {move_data['power']}")
        print(f"Puntos de poder del movimiento: {move_data['pp']}")
        
    else:
        print(f"No se pudo obtener la información del movimiento. Código de respuesta: {move_response.status_code}")
```

# Evolucion a  Tabular

Ahora para cada request (`pokemon` y `movimiento`) convierte su informacion a datos tabulares. Toma en cuenta que vamos a realizar operaciones sobre esos DataFrames, por lo tanto tienes que elegir un formato adecuado.  
Para cada pokemon, y para cada movimiento debe existir un `DataFrame`

```python
pokemon_dataframes = []
moves_dataframes = []

processed_moves = []

for pokemon in pokemons:
    # Datos del Pokémon
    pokemon_url = f'https://pokeapi.co/api/v2/pokemon/{pokemon}'
    pokemon_response = requests.get(pokemon_url)

    if pokemon_response.status_code == 200:
        pokemon_data = pokemon_response.json()
        
        # Obtener información relevante
        nombreP = pokemon_data['name']
        pokemon_id = pokemon_data['id']
        tipos = ', '.join([tipo['type']['name'] for tipo in pokemon_data['types']])
        moves = [{'name': move['move']['name'], 'url': move['move']['url']} for move in pokemon_data['moves']]
        movimientos = ', '.join([move['move']['name'] for move in pokemon_data['moves']])
        moves_df = pd.DataFrame(columns=['Nombre', 'Precisión', 'Tipo', 'Poder', 'PP'])

        for move in moves:
            if move not in processed_moves:
                processed_moves.append(move)
                # Datos del movimiento
                move_response = requests.get(move['url'])

                if move_response.status_code == 200:
                    move_data = move_response.json()

                    # Obtener información relevante
                    nombre = move_data['name']
                    precision = move_data['accuracy']
                    tipo = move_data['type']['name']
                    poder = move_data['power']
                    pp = move_data['pp']

                    # Agregar una fila al DataFrame
                    moves_df = pd.DataFrame({'Nombre': [nombre], 'Precisión': [precision], 'Tipo': [tipo], 'Poder': [poder], 'PP': [pp]},)

                    moves_dataframes.append(moves_df)
        
        # Crear un DataFrame individual para cada Pokémon
        pokemon_df = pd.DataFrame({'Nombre': [nombreP], 'ID': [pokemon_id], 'Tipos': [tipos], 'Movimientos': [movimientos]})
        
        # Agregar el DataFrame individual a la lista
        pokemon_dataframes.append(pokemon_df)
        
    else:
        print(f"No se pudo obtener la información del Pokémon {pokemon}. Código de respuesta: {pokemon_response.status_code}")

# Concatenar todos los DataFrames individuales en uno solo
pokemon_df = pd.concat(pokemon_dataframes, ignore_index=True)
moves_df = pd.concat(moves_dataframes, ignore_index=True)

```
