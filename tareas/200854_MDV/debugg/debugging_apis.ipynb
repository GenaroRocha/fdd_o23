# API Request para varios pokemons

Ahora que sabes como funciona el `API` realiza las peticiones de los siguientes pokemones respecto a sus caracteristicas
y moviemientos, y guardalas en una lista, diciconario o tabla (lo que mas te acomode). 

```python
pokemons =['lucario', 'pikachu', 'gengar', 'charizard', 'blastoise', 'venusaur', 'dragonite', 'mew', 'mewtwo', 'garchomp']
```

```python
pokes = []

for pokemon in pokemons:
    # Datos del Pokémon
    pokemon_url = f'https://pokeapi.co/api/v2/pokemon/{pokemon}'
    
    # Hacer la solicitud para obtener los datos del Pokémon
    res = r.get(pokemon_url)
    
    # Verificar si la solicitud fue exitosa (código 200)
    if res.status_code == 200:
        # Obtener datos del Pokémon en formato JSON
        data = res.json()
        
        # Crear un diccionario para almacenar los datos del Pokémon
        datos_pokemon = {
            'id': data['id'],
            'name': data['name'],
            'types': [t['type']['name'] for t in data['types']],
            'moves': [m['move']['name'] for m in data['moves']]
        }
        
        # Agregar el diccionario a la lista pokes
        pokes.append(datos_pokemon)
    else:
        # Si la solicitud no fue exitosa, imprimir un mensaje de error
        print(f"Error al obtener datos del Pokémon {pokemon}. Código de estado: {res.status_code}")

pokes 




pokes
```

```python
# Tu codigo aqui
moves = df['moves'].explode()
counts = moves.value_counts()
counts = counts[counts >= 3]
counts

MyC = []
''''''
for i in counts.index:
    #recorremos todos los movimientos mayores o iguales a 3
    url = f'https://pokeapi.co/api/v2/move/{i}'
    res = r.get(url)
    moves_data = res.json()
    #guardamos la informacion de los movimientos en un diccionario
    pok_info = {
        'name': moves_data['name'],
        'accuracy': moves_data['accuracy'],
        'type': moves_data['type']['name'],
        'power': moves_data['power'],
        'pp': moves_data['pp'],
        'ocurrences': counts[i]
    }

    MyC.append(pok_info)

df1 = pd.DataFrame(MyC)
df1
    
    


```

# Manejo de tipos

Para cada tipo de pokemon (fire, water, grass, etc) que aparecen en los datos. Obten una serie de dataframes que contengan a los pokemones que son de cada tipo. Y colocalos en un diccionarion donde la llave (`key`) es el tipo

```python
# Tu codigo aqui
tipos = df['types'].explode()
tipos = tipos.value_counts()
df_tipos = {}

for i in tipos.index:
    df_tipos[i] = df[df['types'].apply(lambda x: i in x)]
    
for keys in df_tipos.keys():
    print(df_tipos[keys])

```

Despues guara ese diccionario en un json llamado `tipos.json`

```python
# Tu codigo aqui
# Nonbre del json `pokemon_tipes.json`
import json

for key, df in df_tipos.items():
    df_tipos[key] = df.to_json(orient='records')
    

with open("tipos.json", "w") as json_file:
    json.dump(df_tipos, json_file, indent=2)

```
