Existen varias columnas de fechas. Ninguna tiene un formate de fecha. Para corregir esto haremos un fechanator que moldee las entradas a una forma a la que se le pueda aplicar el formato de fecha.

Ojo. last payment tenia nans. el fechanator los ignora.

```python
#antes 

print(loans['earliest_cr_line'].head(10))
print(loans['last_credit_pull_d'].head(10))
print(loans['last_pymnt_d'].head(10))
print(loans['issue_d'].head(10))
print(loans['settlement_status'].head(10))

print(loans['earliest_cr_line'].isna().any())
print(loans['last_credit_pull_d'].isna().any())
print(loans['last_pymnt_d'].isna().any())#la respuesta que va a hacer que me de de baja es este true
print(loans['issue_d'].isna().any())
print(loans['settlement_status'].isna().any())

loansf['earliest_cr_line'] = loansf['issue_d'].astype(str)
loansf['last_credit_pull_d'] = loansf['last_credit_pull_d'].astype(str)
loansf['last_pymnt_d'] = loansf['last_pymnt_d'].astype(str)
loansf['issue_d'] = loansf['issue_d'].astype(str)
loansf['settlement_status'] = loansf['settlement_status'].astype(str)


def fechanator(cell):
    
    if type(cell) == float: 
        return cell
    if cell=='nan':#odiamos los nan
        return cell
    
    values = cell.split('-')
    key = values[0]
    converter = {
    'Jan': '01',
    'Feb': '02',
    'Mar': '03',
    'Apr': '04',
    'May': '05',
    'Jun': '06',
    'Jul': '07',
    'Aug': '08',
    'Sep': '09',
    'Oct': '10',
    'Nov': '11',
    'Dec': '12'
    }
    return '01-'+converter[key]+'-'+values[1]


loansf['earliest_cr_line'] = loansf['earliest_cr_line'].apply(lambda x: fechanator(x))
loansf['earliest_cr_line'] = pd.to_datetime(loansf['earliest_cr_line'], format= '%d-%m-%Y')

loansf['issue_d'] = loansf['issue_d'].apply(lambda x: fechanator(x))
loansf['issue_d'] = pd.to_datetime(loansf['issue_d'], format= '%d-%m-%Y')

loansf['last_credit_pull_d'] = loansf['last_credit_pull_d'].apply(lambda x: fechanator(x))
loansf['last_credit_pull_d'] = pd.to_datetime(loansf['last_credit_pull_d'], format= '%d-%m-%Y')

loansf['last_pymnt_d'] = loansf['last_pymnt_d'].apply(lambda x: fechanator(x))
loansf['last_pymnt_d'] = pd.to_datetime(loansf['last_pymnt_d'], format= '%d-%m-%Y')

loansf['settlement_date'] = loansf['settlement_date'].apply(lambda x: fechanator(x))
loansf['settlement_date'] = pd.to_datetime(loansf['settlement_date'], format= '%d-%m-%Y')

print(loansf['earliest_cr_line'].head(10))
print(loansf['issue_d'].head(10))
print(loansf['last_credit_pull_d'].head(10))
print(loansf['last_pymnt_d'].head(10))
print(loansf['settlement_date'].head(10))
```

Crea codigo para **guardar** y **cargar** un archivo JSON en el que se guarde la `estrategia` y `valor` que utilizaste para **imputar**. Por ejemplo: Si hay una columna que se llama `columna 3` y utilizaste la estrategia de imputacion de media, y existe otra llamada `columna 4` y  elegiste la palabra 'missing' el JSON debera contener:  
  
 `{'columna 3':{'estrategia':'mean', 'valor':3.4}, 'columna 4':{'estrategia':'identificador', 'valor':'missing'}}`  

 De tal manera que para cada columna que tenga un metodo de imputacion apunte a otro diccionario donde el **key** `estrategia` describa de manera sencilla el metodo, y el **key** `valor` el valor usado. En general:   
 `{'nombre de la columna':{'estrategia':'descripcion de estrategia', 'valor':'valor utilizado'}}`. 
 

De utilizar mas de un metodo puedes anidarlos en una lista  
  `[{...},{...}]`.  

Incluso si la columna utilizada no sufrio imputacion, es necesario que la agregues al JSON.

La idea es que cualquier otra persona pueda cargar el el archivo JSON con tu funcion, entender que hiciste y replicarlo facilmente. No existe solo una respuesta correcta, pero tendras que justificar y explicar tus deciciones.
