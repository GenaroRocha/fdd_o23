# API Request para varios pokemons

Ahora que sabes como funciona el `API` realiza las peticiones de los siguientes pokemones respecto a sus caracteristicas
y moviemientos, y guardalas en una lista, diciconario o tabla (lo que mas te acomode). 

```python
pokemons =['lucario', 'pikachu', 'gengar', 'charizard', 'blastoise', 'venusaur', 'dragonite', 'mew', 'mewtwo', 'garchomp']
```

```python

#creamos un pokedex (lista de pokemones para guardar su info)
pokedex = []
movimientos = []
# pada cada pokemon en la lista
for pokemon in pokemons:
    # solicitamos
    pokemon_url = f'https://pokeapi.co/api/v2/pokemon/{pokemon}'
    pokemon_response = requests.get(pokemon_url)
    pokemon_data = pokemon_response.json()
    
    for move in pokemon_data['moves']:
        moveNom = move['move']['name']
        moveUrl = move['move']['url']
        #cada movimiento es agragado a la lista
        movimientos.append({"name": moveNom, "url": moveUrl})
        
    pokemon_dict = {
        "ID": pokemon_data['id'],
        "Nombre ":pokemon_data['name'],
        "Tipo(s)": [", ".join(tipo['type']['name'] for tipo in pokemon_data['types'])],
        "moves": movimientos
    }
    pokedex.append(pokemon_dict)
lista_movimientos = []
#ahora para los moves
for m in movimientos:
        moves_response = requests.get(f'{m.get("url")}')
        moves_data = moves_response.json()
        info_moves = {
            'nombre': moves_data['name'],
            'ac': moves_data['accuracy'],
            'tipo': moves_data['type']['name'],
            'po': moves_data['power'],
            'pp': moves_data['pp']
        }
        lista_movimientos.append(info_moves)
print(pokedex)
print(lista_movimientos)
```

# Evolucion a  Tabular

Ahora para cada request (`pokemon` y `movimiento`) convierte su informacion a datos tabulares. Toma en cuenta que vamos a realizar operaciones sobre esos DataFrames, por lo tanto tienes que elegir un formato adecuado.  
Para cada pokemon, y para cada movimiento debe existir un `DataFrame`

```python

pokemon_df = pd.DataFrame(pokedex)
moves_df = pd.DataFrame(lista_movimientos)

# clean clean
pokemon_df = pokemon_df.explode('Tipo(s)')
pokemon_df = pokemon_df.explode('moves')

def get_name(moves_dict):
    for move_key, move_value in moves_dict.items():
        return move_value

#transformamos la columna moves
pokemon_df['nomMove'] = pokemon_df['moves'].apply(get_name)

pokemon_df = pokemon_df.drop('moves', axis = 1)

# more clean clean
moves_df = moves_df.dropna()
moves_df = moves_df.drop_duplicates()
pokemon_df
```
