# API request

Crea codigo para acceder al API de pokemon y hacer requests. esta primera parte del codigo solo es para entender como es la estructura de los datos

Compelta el siguiente snippet para obter los datos del API de pokemon. La idea de estos primeros snippets es solo entender como funcionan los datos y como llegan. Para ello solo analizamos un pokemon y sus movimientos, mas adelante tenemos que repetirlo pero para varios pokemons.

## Pokemon Data

Para  el tipo de pokemon debes obtener:
1. `pokemon_id`
2. `pokemon_name`
3. `type`(s) cada pokemon tiene a 1 o dos tipos (no mas). Tienes que guardar ambos  

Guarda los `moves` del pokemon pues los usaremos mas adelante. En especial el `name` y la `url` de cada movimiento para que podamos iterar despues

```python
# Elige tu pokemon
pokemon_name = 'lucario'
```

```python
# Solicitud para obtener datos de un Pokémon
pokemon_url = f'https://pokeapi.co/api/v2/pokemon/{pokemon_name}'
pokemon_response = requests.get(pokemon_url)
#pokemon_data = # Completa el formateo de datos
#moves  = # Completa
if pokemon_response.status_code == 200:
    # Formatear los datos en formato JSON
    pokemon_data = pokemon_response.json()

    # Obtener ID y nombre del Pokémon
    pokemon_id = pokemon_data['id']
    pokemon_name = pokemon_data['name']

    # Obtener tipo(s) del Pokémon
    pokemon_types = [type_data['type']['name'] for type_data in pokemon_data['types']]

    # Imprimir información del Pokémon
    print(f"ID del Pokémon: {pokemon_id}")
    print(f"Nombre del Pokémon: {pokemon_name}")
    print(f"Tipo(s) del Pokémon: {pokemon_types}")

    # Obtener movimientos del Pokémon
    moves = [{'name': move['move']['name'], 'url': move['move']['url']} for move in pokemon_data['moves']]

    # Imprimir información de los movimientos
    print("Moves del Pokémon:")
    for move in moves:
        print(f"Nombre: {move['name']}, URL: {move['url']}")

else:
    print(f"No se pudo obtener datos del Pokémon. Código de estado: {pokemon_response.status_code}")
```

## Moves Data

Ahora para cada movimiento `move` del pokemon debes obtener los siguientes datos usando la `url`
1. `name`
2. `accuracy`
3. `type`
4. `power`
5. `pp`

```python
for m in moves:
    moves_response = requests.get(f'{m.get("url")}')
    moves_data = moves_response.json()
    info_moves = {
        'name': moves_data['name'],
        'accuracy': moves_data['accuracy'],
        'type': moves_data['type']['name'],
        'power': moves_data['power'],
        'pp': moves_data['pp']
    }
    print(info_moves)
```

# API Request para varios pokemons

Ahora que sabes como funciona el `API` realiza las peticiones de los siguientes pokemones respecto a sus caracteristicas
y moviemientos, y guardalas en una lista, diciconario o tabla (lo que mas te acomode).

```python
pokemons =['lucario', 'pikachu', 'gengar', 'charizard', 'blastoise', 'venusaur', 'dragonite', 'mew', 'mewtwo', 'garchomp']
```

```python
lista_pokemones = []
movimientos = []

# Solicitudes para obtener datos de un Pokémon
for pokemon in pokemons:
    # Datos del Pokémon
    pokemon_url = f'https://pokeapi.co/api/v2/pokemon/{pokemon}'
    pokemon_response = requests.get(pokemon_url)
    pokemon_data = pokemon_response.json()

    for move in pokemon_data['moves']:
        move_name = move['move']['name']
        move_url = move['move']['url']
        movimientos.append({"name": move_name, "url": move_url})

    pokemon_dict = {
        "pokemon_id": pokemon_data['id'],
        "pokemon_name": pokemon_data['name'],
        "type": [", ".join(tipo['type']['name'] for tipo in pokemon_data['types'])],
        "moves": movimientos
    }
    lista_pokemones.append(pokemon_dict)

# Limpiar la lista de movimientos
for m in moves:
    moves_response = requests.get(f'{m.get("url")}')
    moves_data = moves_response.json()
    info_moves = {
        'name': moves_data['name'],
        'accuracy': moves_data['accuracy'],
        'type': moves_data['type']['name'],
        'power': moves_data['power'],
        'pp': moves_data['pp']
    }
    lista_movimientos.append(info_moves)

print(lista_pokemones)
print(lista_movimientos)
```

# Evolucion a  Tabular

Ahora para cada request (`pokemon` y `movimiento`) convierte su informacion a datos tabulares. Toma en cuenta que vamos a realizar operaciones sobre esos DataFrames, por lo tanto tienes que elegir un formato adecuado.  
Para cada pokemon, y para cada movimiento debe existir un `DataFrame` o puedes guardarlos en uno solo, aunque es recomendable dejarlos por separado.

```python


def obtener_informacion_pokemon(pokemon_name):
    pokemon_url = f'https://pokeapi.co/api/v2/pokemon/{pokemon_name}'
    response = requests.get(pokemon_url)
    if response.status_code == 200:
        pokemon_data = response.json()
        return {
            'Nombre': pokemon_data['name'],
            'ID': pokemon_data['id'],
            'Tipo': [tipo['type']['name'] for tipo in pokemon_data['types']],
            'Moves': [{'name': move['move']['name'], 'url': move['move']['url']} for move in pokemon_data['moves']]
        }
    else:
        return None

def obtener_informacion_movimiento(move_url):
    response = requests.get(move_url)
    if response.status_code == 200:
        move_data = response.json()
        return {
            'Nombre': move_data['name'],
            'Accuracy': move_data.get('accuracy', None),
            'Tipo': move_data['type']['name'],
            'Poder': move_data.get('power', None),
            'PP': move_data.get('pp', None)
        }
    else:
        return None


```

```python
import pandas as pd

# Almacenar la información de cada Pokémon en una lista
informacion_pokemons = []

# Solicitar información para cada Pokémon en la lista
for pokemon in pokemons:
    pokemon_info = obtener_informacion_pokemon(pokemon)
    if pokemon_info:
        informacion_pokemons.append(pokemon_info)

# Crear DataFrame para la información de Pokémon
df_pokemons = pd.DataFrame(informacion_pokemons)

# Almacenar la información de cada movimiento en una lista
informacion_movimientos = []

# Solicitar información para cada movimiento del Pokémon
for _, move in df_pokemons[['Nombre', 'Moves']].explode('Moves').iterrows():
    movimiento_info = obtener_informacion_movimiento(move['Moves']['url'])
    if movimiento_info:
        informacion_movimientos.append({
            'Pokemon': move['Nombre'],
            'Move': movimiento_info['Nombre'],
            'Accuracy': movimiento_info['Accuracy'],
            'Tipo': movimiento_info['Tipo'],
            'Poder': movimiento_info['Poder'],
            'PP': movimiento_info['PP']
        })

# Crear DataFrame para la información de movimientos
df_movimientos = pd.DataFrame(informacion_movimientos)

# Mostrar DataFrames
print("DataFrame de Pokémon:")
print(df_pokemons)

print("\nDataFrame de Movimientos:")
print(df_movimientos)
```

# Manejo de tipos

Para cada tipo de pokemon (fire, water, grass, etc) que aparecen en los datos. Obten una serie de dataframes que contengan a los pokemones que son de cada tipo. Y colocalos en un diccionarion donde la llave (`key`) es el tipo

```python
# Asumiendo que ya tienes el DataFrame df_pokemons creado

# Crear un diccionario para almacenar los DataFrames de cada tipo
df_tipos = {}

# Obtener tipos únicos de Pokémon
tipos_unicos = df_pokemons['Tipo'].explode().unique()

# Iterar sobre cada tipo y filtrar los datos
for tipo in tipos_unicos:
    # Filtrar el DataFrame para obtener solo los Pokémon de este tipo
    df_tipo_actual = df_pokemons[df_pokemons['Tipo'].apply(lambda tipos: tipo in tipos)]

    # Almacenar el DataFrame en el diccionario
    df_tipos[tipo] = df_tipo_actual

# Mostrar el diccionario resultante
print("Diccionario de DataFrames por Tipo:")
print(df_tipos)
```

Despues guara ese diccionario en un json llamado `tipos.json`

```python
import json

# Asumiendo que ya tienes el diccionario df_tipos creado

# Convertir cada DataFrame a un diccionario
df_tipos_serializable = {tipo: df_tipo.to_dict(orient='records') for tipo, df_tipo in df_tipos.items()}

# Nombre del archivo JSON
nombre_archivo = 'pokemon_tipes.json'

# Guardar el diccionario en el archivo JSON
with open(nombre_archivo, 'w') as archivo:
    json.dump(df_tipos_serializable, archivo)

print(f"El diccionario se ha guardado en el archivo {nombre_archivo}")
```

Ahora crea un DataFrame que para cada pokemon unicamente contenga los movimientos que son de su tipo (osea si es tipo fuego y voldaor solo usa movimientos de tipo volador y fuego). Y crea una tabla donde tengan los siguientes datos en formato `tidydata`:
1. `pokemon_id`
2. `pokemon_name`
3. `type`(s) cada pokemon tiene a 1 o dos tipos (no mas). Tienes que guardar ambos  
4. `name` (del movimiento)
5. `accuracy`
6. `type`
7. `power`
8. `pp`

```python
print(df_pokemons.head())
```

```python
# Inicializar listas para almacenar datos en formato tidydata
tidydata_rows = []

# Iterar sobre cada Pokémon y sus movimientos
for index, row in df_pokemons.iterrows():
    pokemon_id = row['ID']
    pokemon_name = row['Nombre']
    tipos = row['Tipo']

    # Filtrar movimientos por tipo de Pokémon
    movimientos_pokemon = df_movimientos[df_movimientos['Tipo'].isin(tipos)]

    # Iterar sobre los movimientos del Pokémon
    for index_move, row_move in movimientos_pokemon.iterrows():
        move_data = {
            'pokemon_id': pokemon_id,
            'pokemon_name': pokemon_name,
            'type': tipos,
            'name': row_move['Move'],
            'accuracy': row_move['Accuracy'],
            'move_type': row_move['Tipo'],
            'power': row_move['Poder'],
            'pp': row_move['PP'],
        }
        tidydata_rows.append(move_data)

# Crear el DataFrame en formato tidydata
df_tidydata = pd.DataFrame(tidydata_rows)

# Mostrar el DataFrame resultante
print("DataFrame en formato tidydata:")
print(df_tidydata)

# Guardar el DataFrame en un archivo CSV
df_tidydata.to_csv('tidydata_pokemon_moves.csv', index=False)
```
